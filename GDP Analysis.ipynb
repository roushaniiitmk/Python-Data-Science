## Get the names of the states/union territories present in the list. Will calculate mean GDP for these states

```python
# Get the names of columns which has names of states/union territories and remove ('Items Description','Duration') 
column_attributes = list(gdp_current_price.columns.values)
list_remove = ('Items Description','Duration')  # Do not calculate mean for the these 2 columns
states = [ val for val in column_attributes if val not in list_remove] 
print(states)
```

## Loop through the states to calculate the average gdp growth of each state over the period 2013-14, 2014-15, 2015-16. States are present as column.

```python
state_mean = {}
for column in gdp_current_price:
    # ignore the ('Items Description','Duration')
    if column in states:
        # calculate the mean of the 
        mean=gdp_current_price[gdp_current_price['Items Description']=='(% Growth over previous year)'][column].mean()
        # create a dictionary to store all the mean value of all the states
        state_mean[column]=mean
        
# Append state mean dictionary to the column 'Items Description' under the column name 'Average growth rate'
state_mean['Items Description']='Average growth rate'
gdp_current_price = gdp_current_price.append(state_mean, ignore_index=True)
```

```python
# 'Average growth rate' is added under the category 'Item Description'
gdp_current_price
```

```python
# Drop (% Growth over previous year) for the year '2013-14'
index_to_drop = gdp_current_price.index[((gdp_current_price['Items Description']=='(% Growth over previous year)') & (gdp_current_price['Duration']=='2013-14'))]
gdp_current_price = gdp_current_price.drop(index_to_drop)

# Drop (% Growth over previous year) for the year '2014-15'
index_to_drop = gdp_current_price.index[((gdp_current_price['Items Description']=='(% Growth over previous year)') & (gdp_current_price['Duration']=='2014-15'))]
gdp_current_price = gdp_current_price.drop(index_to_drop)


# Drop (GSDP - CURRENT PRICES (` in Crore)) for the year '2011-12'
index_to_drop =  gdp_current_price.index[((gdp_current_price['Items Description']=='GSDP - CURRENT PRICES (` in Crore)') & (gdp_current_price['Duration']=='2011-12'))]
gdp_current_price = gdp_current_price.drop(index_to_drop)

# Drop (GSDP - CURRENT PRICES (` in Crore)) for the year '2012-12'
index_to_drop =  gdp_current_price.index[((gdp_current_price['Items Description']=='GSDP - CURRENT PRICES (` in Crore)') & (gdp_current_price['Duration']=='2012-13'))]
gdp_current_price = gdp_current_price.drop(index_to_drop)

# Drop (GSDP - CURRENT PRICES (` in Crore)) for the year '2013-14'
index_to_drop =  gdp_current_price.index[((gdp_current_price['Items Description']=='GSDP - CURRENT PRICES (` in Crore)') & (gdp_current_price['Duration']=='2013-14'))]
gdp_current_price = gdp_current_price.drop(index_to_drop)

# Drop (GSDP - CURRENT PRICES (` in Crore)) for the year '2014-15'
index_to_drop =  gdp_current_price.index[((gdp_current_price['Items Description']=='GSDP - CURRENT PRICES (` in Crore)') & (gdp_current_price['Duration']=='2014-15'))]
gdp_current_price = gdp_current_price.drop(index_to_drop)

# Remove Duration 
gdp_current_price = gdp_current_price.drop('Duration', axis = 1)
```

## Finding the depedency of GDP on primary, secondary and tertiary.

```python
# creating a time slot for various time period of day
def gdp_percentage(gsdp, total_gsdp):
    return gsdp/total_gsdp
    
gsva['primary_gdp_percent'] = gsva[['Primary','Gross State Domestic Product']].apply(lambda x: gdp_percentage(*x), axis=1)
gsva['secondary_gdp_percent'] = gsva[['Secondary','Gross State Domestic Product']].apply(lambda x: gdp_percentage(*x), axis=1)
gsva['tertiary_gdp_percent'] = gsva[['Tertiary','Gross State Domestic Product']].apply(lambda x: gdp_percentage(*x), axis=1)
```

If better infrastructure is developed to store the goods, farmers can get better prices for their crops as they do not need to sell their products at cheaper prices if the productivity is high.
